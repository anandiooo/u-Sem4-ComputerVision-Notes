```python
import cv2 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./assets/lenna.png", cv2.IMREAD_GRAYSCALE) # -> convert langsung di imread

# cv.imshow() -> beda

# cara 1 -> cvtColor
# img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

print(img.shape)
print(img)

plt.imshow(img, cmap="grey") # di notebook lebih enak pake plt
plt.axis(False)
plt.show()
```

```python
# threshold
_, bin = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)
_, invBin = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY_INV)
_, trunc = cv2.threshold(img, 100, 255, cv2.THRESH_TRUNC)
_, tozero = cv2.threshold(img, 100, 255, cv2.THRESH_TOZERO)
_, invToZero = cv2.threshold(img, 100, 255, cv2.THRESH_TOZERO_INV)
_, otsu = cv2.threshold(img, 100, 255, cv2.THRESH_OTSU)

thresImages = {
    "Original" : img,
    "Truncate" : trunc,
    "Binary" : bin,
    "InvBinary" : invBin,
    "ToZero" : tozero,
    "InvToZero" : invToZero,
    "Otsu" : otsu
}

plt.figure(figsize=(16,12))
for i, (currTitle, currImg) in enumerate(thresImages.items()):
    plt.subplot(4,2,i+1)
    plt.imshow(currImg, cmap="grey")
    plt.title(currTitle)
    plt.axis(False)

# print(_)
# plt.imshow(bin, cmap="grey")
# plt.axis(False)
# plt.show()
```

```python
# adaptive thresholding
sudoku = cv2.imread("./assets/sudoku.jpg", cv2.IMREAD_GRAYSCALE)

_, sudokiBin = cv2.threshold(sudoku, 120, 255, cv2.THRESH_OTSU)
sudokuMean = cv2.adaptiveThreshold(sudoku, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11,2)
sudokuGaussian = cv2.adaptiveThreshold(sudoku, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11,2)

adaptImages = {
    "Original" : sudoku,
    "Sudoku Bin" : sudokiBin,
    "Sudoku Mean" : sudokuMean,
    "Sudoku Gaussian" : sudokuGaussian
}

plt.figure(figsize=(16,12))
for i, (currTitle, currImg) in enumerate(adaptImages.items()):
    plt.subplot(2,2,i+1)
    plt.imshow(currImg, cmap="grey")
    plt.title(currTitle)
    plt.axis(False)
```

```python
calvin = cv2.imread("./assets/chaplin.png", cv2.IMREAD_GRAYSCALE)

calvinMeanBlur = cv2.blur(calvin, (7,7))
calvinMedianBlur = cv2.medianBlur(calvin, 7)
calvainGaussianBlur = cv2.GaussianBlur(calvin, (7,7), 5)
calvinBilateralFilter = cv2.bilateralFilter(calvin, 9, 75, 75)

blurImages = {
    "Original" : calvin,
    "mean blur" : calvinMeanBlur,
    "median blur" : calvinMedianBlur,
    "gaussian" : calvainGaussianBlur,
    "bilateral" : calvinBilateralFilter
}

plt.figure(figsize=(20,12))
for i, (currTitle, currImg) in enumerate(blurImages.items()):
    plt.subplot(1,5,i+1)
    plt.imshow(currImg, cmap="grey")
    plt.title(currTitle)
    plt.axis(False) 

plt.imshow(calvinMeanBlur, cmap="grey")
```
